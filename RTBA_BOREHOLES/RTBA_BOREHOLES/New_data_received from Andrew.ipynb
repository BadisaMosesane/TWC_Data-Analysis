## Cleaning datasets

- Missing dates in columns: compltdate and swl_date. Missing data labelled NO_DATA
- Repalace 'No-Data' with 'unknown' 
- Fill in data owner, contact persion, email and file name columns 
- Rename columns using appropriate RIMS template column names

```python
bots_clean = Botswana_Bhs.copy()
south_clean = SouthAfrica_Bhs.copy()
```

```python
bots_clean ['dataowner'] = "SADC-GIP"
bots_clean ['contact_person'] = "Brighton Munyai"
bots_clean ['email'] = "Brighton@sadc-gmi.org"
bots_clean ['file_name'] = "ramotsw_TBA_Botswana_boreholes.dbf"
```

```python
south_clean ['dataowner'] = "SADC-GIP"
south_clean ['contact_person'] = "Brighton Munyai"
south_clean ['email'] = "Brighton@sadc-gmi.org"
south_clean ['file_name'] = "ramotsw_TBA_Botswana_boreholes.dbf"
```

```python
gwd = pd.concat([bots_clean, south_clean], sort = False)

gwd = gwd.rename(columns={'natl_bhid':'bh_id',
                    'sadc_bhid':'bh_id_sadc',
                    'long_wgs84':'longitude',
                    'lat_wgs84':'latitude',
                    'bh_depth_m':'bh_depth',
                    'elev_m_sub':'elevation',
                    'compltdate':'compl_date',
                    'activ_stat':'bh_status',
                    'lith_subm':'lith'});
```

```python
gwd.replace('Unknown', 'unknown', inplace = True)
gwd.replace('NO_DATA', 'unknown', inplace = True)
print (gwd ['yieldtype'].unique())
```

```python
gwd ['compl_date'] = pd.to_datetime(gwd ['compl_date'], yearfirst = True, errors = 'coerce')
gwd ['swl_date'] = pd.to_datetime(gwd ['swl_date'], yearfirst = True, errors = 'coerce')
gwd ['compl_date'].dt.strftime('%Y-%m-%d')
gwd ['swl_date'].dt.strftime('%Y-%m-%d');
```

```python
gwd.drop (['compltd_yr'], axis = 1, inplace = True)
```
